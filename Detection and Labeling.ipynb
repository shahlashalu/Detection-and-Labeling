```python
# Input and Output video files
input_video_path = "C:/Users/hp/objdtctn.mp4"
output_video_path = "C:/Users/hp/detected_video_label_det.avi"
```

```python
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained = True)
```

```python
cap = cv2.VideoCapture(input_video_path)
fps = cap.get(cv2.CAP_PROP_FPS)
frame_width, frame_height = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)), int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
print(f"FPS: {fps}, frame width : {frame_width}, frame height : {frame_height}")
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

#Process each frame

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Perform object detection
    results = model(frame)   # YOLOv5 inference
    #print('results', results)

    # Extract detections
    detections = results.pandas().xyxy[0]   # Bounding box, class, and confidence
    print('detections', detections)
    for _, detection in detections.iterrows():
        x1, y1, x2, y2 = int(detection['xmin']), int(detection['ymin']), int(detection['xmax']), int(detection['ymax'])
        confidance = detection['confidence']
        label = f"{detection['name']} {confidance:.2f}"

        # Draw bounding box and label
        color = (0, 255, 0)   # Green color
        cv2.rectangle(frame, (x1,y1), (x2,y2), color, 2)
        cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)


    # Write frame to the output video inside the loop
    print('ready to write')
    out.write(frame)
    #print('frame written output video')
    

# Release resources
cap.release()
out.release()
cv2.destroyAllWindows()
print(f"Video with detected faces saved as (output_video)")
```
